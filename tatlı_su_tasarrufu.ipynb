```python
import pandas as pd
import matplotlib.pyplot as plt

# Su tasarrufu verisi
data = {
    "Yıl": [2023, 2022, 2021, 2020, 2019],
    "Tatlı Su Tasarrufu (milyon galon)": [12700, 13000, 12300, 10800, 9300]
}

df = pd.DataFrame(data)
df.set_index("Yıl", inplace=True)

# Grafik oluştur
plt.figure(figsize=(10, 7))
plt.plot(df.index, df["Tatlı Su Tasarrufu (milyon galon)"], marker='o', color="#5E81AC", linewidth=2)
plt.fill_between(df.index, df["Tatlı Su Tasarrufu (milyon galon)"], color="#5E81AC", alpha=0.2)

# Değerleri göster
for year, value in zip(df.index, df["Tatlı Su Tasarrufu (milyon galon)"]):
    plt.annotate(f'{value:,}', (year, value), textcoords="offset points", xytext=(0,10), ha='center')

# Başlık ve etiketler
plt.title("Apple'ın Yıllara Göre Tatlı Su Tasarrufu (2019–2023)", fontsize=14)
plt.xlabel("Yıl", fontsize=12)
plt.ylabel("Milyon Galon", fontsize=12)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

Apple’ın Tatlı Su Tasarrufu Performansı (2019–2023)
Apple’ın 2019–2023 dönemine ait tatlı su tasarrufu verileri incelendiğinde, şirketin operasyonel ve tedarik zinciri düzeyinde uyguladığı su verimliliği politikalarının somut etkiler yarattığı görülmektedir. 2019 yılında yaklaşık 9.3 milyon galon olan tasarruf miktarı, 2022 yılında 13 milyon galon seviyesine çıkarak %40’ın üzerinde bir artış göstermiştir. 2023 yılında ise bu miktar 12.7 milyon galon ile sınırlı bir düşüş gösterse de, genel eğilim açısından yüksek seviyeler korunmuştur.

Bu eğilim, Apple’ın su stresinin yüksek olduğu bölgelerde yürüttüğü Tedarikçi Temiz Su Programı kapsamında yapılan yatırımlar ve üretim tesislerinde uygulanan gri su arıtma sistemleri, soğutma verimliliği iyileştirmeleri gibi teknolojik önlemlerin etkili biçimde devreye alındığını ortaya koymaktadır.

Değerlendirme:
Apple, su kaynaklarının korunmasına yönelik olarak geliştirdiği stratejik programlar ve veri merkezlerinden üretim tesislerine uzanan geniş kapsamlı uygulamalar sayesinde yıllık milyonlarca galon tatlı su tasarrufu sağlamaktadır. Bu durum, şirketin yalnızca karbon ayak izini değil, aynı zamanda su ayak izini azaltmaya yönelik bütüncül sürdürülebilirlik yaklaşımını da yansıtmaktadır.

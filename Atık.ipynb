```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Veriyi DataFrame olarak oluşturma
data = {
    'Atık Türü': ['Çöp Sahasına Atılmış', 'Çöp Sahasına Atılmış', 'Çöp Sahasına Atılmış', 'Çöp Sahasına Atılmış', 'Çöp Sahasına Atılmış',
                  'Geri Dönüştürülmüş', 'Geri Dönüştürülmüş', 'Geri Dönüştürülmüş', 'Geri Dönüştürülmüş', 'Geri Dönüştürülmüş',
                  'Kompostlanmış', 'Kompostlanmış', 'Kompostlanmış', 'Kompostlanmış', 'Kompostlanmış',
                  'Tehlikeli Atık', 'Tehlikeli Atık', 'Tehlikeli Atık', 'Tehlikeli Atık', 'Tehlikeli Atık'],
    'Yıl': [2023, 2022, 2021, 2020, 2019,
            2023, 2022, 2021, 2020, 2019,
            2023, 2022, 2021, 2020, 2019,
            2023, 2022, 2021, 2020, 2019],
    'Miktar (Pounds)': [38343490, 33260990, 33202200, 25826550, 38317120,
                        81781660, 78618250, 73489220, 63812300, 72338130,
                        14803510, 8726170, 4844960, 6302410, 10882110,
                        7321130, 2780610, 3525840, 4053770, 6096000]
}
df = pd.DataFrame(data)

# 'Yıl' sütununu sıralama
df = df.sort_values(by=['Atık Türü', 'Yıl'])

print(df.head())
```

1. Yıllara Göre Atık Türlerinin Miktar Değişimini Gösteren Çizgi Grafiği
Bu grafik, her bir atık türünün zaman içindeki eğilimini anlamak için idealdir. Farklı atık türlerini ayrı çizgilerle göstererek karşılaştırma yapmayı kolaylaştırır.
Eğilimleri Gösterme: Her atık türünün zaman içindeki artış, azalış veya sabit kalma eğilimini net bir şekilde gösterir.
Karşılaştırma: Farklı atık türlerinin performansını aynı anda karşılaştırma imkanı sunar. Örneğin, geri dönüştürülmüş atık miktarının sürekli arttığını, çöp sahasına atılan atığın ise dalgalandığını görebilirsiniz.

```python
plt.figure(figsize=(12, 7))
sns.lineplot(data=df, x='Yıl', y='Miktar (Pounds)', hue='Atık Türü', marker='o')
plt.title('Yıllara Göre Atık Türlerinin Miktar Değişimi', fontsize=16)
plt.xlabel('Yıl', fontsize=12)
plt.ylabel('Miktar (Pounds)', fontsize=12)
plt.xticks(df['Yıl'].unique().astype(int)) # Yıl eksenindeki etiketleri tam sayı olarak göster
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Atık Türü', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

2. Yıllara Göre Toplam Atık Miktarlarının Yığılmış Çubuk Grafiği
Bu grafik, her yıl üretilen toplam atık miktarının nasıl dağıldığını ve her atık türünün bu toplama ne kadar katkıda bulunduğunu görselleştirir.
Toplam Katkı: Her atık türünün yıllık toplam atık içindeki payını görselleştirir.
Yıllık Kompozisyon: Yıllar geçtikçe atık kompozisyonunda meydana gelen değişiklikleri (örneğin, geri dönüşüm payının artması) anlamaya yardımcı olur.
Yığılmış Görünüm: Toplam atık miktarının her yıl ne kadar olduğunu ve bu toplamın nasıl bölündüğünü tek bir grafikte gösterir.

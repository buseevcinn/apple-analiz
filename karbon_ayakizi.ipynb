```python
import pandas as pd
import matplotlib.pyplot as plt

# Karbon ayak izi verisi
data = {
    "Yıl": [2023, 2022, 2021, 2020, 2019],
    "Toplam Net Karbon Ayak İzi (ton CO2e)": [15600000, 20300000, 22530000, 22530000, 25100000]
}

df = pd.DataFrame(data)
df.set_index("Yıl", inplace=True)
plt.figure(figsize=(10, 6))
plt.plot(df.index, df["Toplam Net Karbon Ayak İzi (ton CO2e)"], marker='o', color="#BF616A", linewidth=2)
plt.fill_between(df.index, df["Toplam Net Karbon Ayak İzi (ton CO2e)"], color="#BF616A", alpha=0.2)
for year, value in zip(df.index, df["Toplam Net Karbon Ayak İzi (ton CO2e)"]):
    plt.annotate(f'{value/1e6:.2f}M', (year, value), textcoords="offset points", xytext=(0,10), ha='center')
plt.title("Apple'ın Toplam Net Karbon Ayak İzi (2019–2023)", fontsize=14)
plt.xlabel("Yıl", fontsize=12)
plt.ylabel("Ton CO2e", fontsize=12)
plt.grid(alpha=0.6)
plt.tight_layout()
plt.show()
```

Apple’ın 2019–2023 Dönemi Net Karbon Ayak İzi Eğilimi
Apple’ın 2019–2023 yılları arasındaki toplam net karbon ayak izi (metrik ton CO₂e) verileri incelendiğinde, şirketin sürdürülebilirlik politikalarının etkili bir şekilde uygulandığı ve emisyon azaltım stratejilerinin ölçülebilir sonuçlar doğurduğu açıkça görülmektedir. 2019 yılında 25,1 milyon ton CO₂e olan karbon ayak izi, 2023 yılı itibarıyla 15,6 milyon ton CO₂e seviyesine gerileyerek yaklaşık %38’lik bir azalma göstermiştir.

Bu azalma, özellikle üretim süreçlerinde yenilenebilir enerji kullanımının yaygınlaştırılması, tedarik zinciri optimizasyonu ve ürünlerin enerji verimliliğinde sağlanan ilerlemelerle ilişkilendirilebilir. 2020 ve 2021 yıllarında karbon ayak izinin aynı seviyede kalması (22,53 milyon ton CO₂e), COVID-19 pandemisinin etkileri ve küresel tedarik zinciri dalgalanmalarının etkisiyle durağan bir dönem yaşandığını göstermektedir.

2022 yılında geçici bir artış gözlense de, 2023’te gerçekleşen rekor düzeydeki emisyon azalımı, Apple’ın “2030 karbon nötr” hedefine doğru ivmelenerek ilerlediğini kanıtlamaktadır.

Değerlendirme:
Apple’ın karbon ayak izinde 5 yıllık dönemde sağlanan bu sistematik azalma, şirketin çevresel sürdürülebilirlik alanındaki stratejik liderliğini güçlendirmekte ve teknoloji sektörüne öncülük eden bir emisyon yönetimi modeli sunduğunu göstermektedir.

1- Zaman Serisi Çizgi Grafiği (Trend Analizi) (Enerji tiplerinin yıllara göre tüketimini görselleştirmek için)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Veri çerçevesi
data = {
    "Yıl": [2019, 2020, 2021, 2022, 2023]*3,
    "Enerji Tipi": ["Elektrik"]*5 + ["Yakıt"]*5 + ["Buhar, Isıtma ve Soğutma"]*5,
    "Tüketim (MWh)": [
        2427000, 2580000, 2854000, 3199000, 3487000,
        462680, 439170, 467280, 334250, 662950,
        0, 0, 22480, 19800, 45370
    ]
}
df = pd.DataFrame(data)

# Çizgi grafiği
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="Yıl", y="Tüketim (MWh)", hue="Enerji Tipi", marker="o")
plt.title("Yıllara Göre Enerji Tüketimi")
plt.ylabel("Tüketim (MWh)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

 Alan (Stackplot) Grafiği (Zaman içinde enerji tipi dağılımını alanlarla göstermek için)

```python
pivot_df = df.pivot(index="Yıl", columns="Enerji Tipi", values="Tüketim (MWh)").fillna(0)
pivot_df.plot.area(figsize=(10, 6), cmap="Set2")
plt.title("Enerji Tüketim Dağılımı (Alan Grafiği)")
plt.ylabel("Tüketim (MWh)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Genel Değerlendirme: “2023 yılında Apple’ın enerji tüketimi tüm kategorilerde artış göstermiştir. En yüksek artış elektrik tüketiminde görülmektedir.”

Buhar, Isıtma ve Soğutma: “2019-2020’de bu alanda sıfır tüketim raporlanmışken, 2021’den itibaren artış dikkat çekicidir.”

Yakıt Kullanımı: “2022 yılında dramatik bir düşüş görülse de, 2023’te neredeyse iki kat artmıştır.”

Sürdürülebilirlik Yorumu: “Elektrik tüketimindeki artış yenilenebilir kaynaklardan sağlandıysa olumlu; ancak fosil yakıt kullanımı yeniden artışa geçmiş görünüyor.”

```python
import pandas as pd
import matplotlib.pyplot as plt

# Enerji ve emisyon verileri
data = {
    "Mali Yıl": [
        2011, 2012, 2013, 2014, 2015, 2016, 2017,
        2018, 2019, 2020, 2021, 2022, 2023
    ],
    "Toplam Enerji Kullanımı (kWh)": [
        42500, 38552300, 79462900, 108659700, 142615000, 145520900, 289195800,
        327663800, 339047649, 372901398, 384727076, 487921930, 483299062
    ],
    "Yenilenebilir Enerji Kullanımı (%)": [
        0, 4, 59, 81, 85, 98, 99, 100, 100, 100, 100, 100, 100
    ],
    "Apple'ın Emisyonları (ton CO2e)": [
        10, 16500, 14500, 11000, 12700, 1600, 1500, 400, 0, 0, 0, 0, 0
    ]
}

df = pd.DataFrame(data)
df.set_index("Mali Yıl", inplace=True)

# Grafik oluştur
fig, ax1 = plt.subplots(figsize=(12, 6))

# Enerji kullanımını çubuk grafikle göster
ax1.bar(df.index, df["Toplam Enerji Kullanımı (kWh)"], color='#A3BE8C', label='Toplam Enerji Kullanımı (kWh)')
ax1.set_ylabel("Toplam Enerji (kWh)", color='#A3BE8C')
ax1.tick_params(axis='y', labelcolor='#A3BE8C')

# İkincil y eksenine emisyonları ekle
ax2 = ax1.twinx()
ax2.plot(df.index, df["Apple'ın Emisyonları (ton CO2e)"], color='#BF616A', marker='o', linewidth=2, label="Apple'ın Emisyonları (ton CO2e)")
ax2.set_ylabel("Apple'ın Emisyonları (ton CO2e)", color='#BF616A')
ax2.tick_params(axis='y', labelcolor='#BF616A')

# Başlık ve gösterge
plt.title("Apple'ın Enerji Kullanımı ve Emisyon Eğilimi (2011–2023)", fontsize=14)
fig.legend(loc='upper left', bbox_to_anchor=(0.15, 0.85))
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

Apple’ın Enerji Kullanımı ve Emisyon Eğilimleri (2011–2023)
Apple’ın son 13 yıllık enerji tüketimi ve emisyon verileri incelendiğinde, şirketin yenilenebilir enerjiye geçiş sürecinin yalnızca enerji kaynaklarını değiştirmekle kalmayıp, aynı zamanda karbon emisyonlarını da neredeyse tamamen ortadan kaldıracak şekilde sistematik bir dönüşüm içerdiği görülmektedir.

Toplam Enerji Kullanımı:
2011 yılında yalnızca 42.500 kWh seviyesinde olan toplam enerji tüketimi, 2023 itibarıyla yaklaşık 483 milyon kWh düzeyine yükselmiştir. Bu artış, Apple’ın küresel operasyonlarının büyüklüğünü ve enerjiye olan ihtiyacını açıkça ortaya koymaktadır. Ancak bu büyümeye rağmen karbon emisyonlarındaki dramatik düşüş, şirketin enerji yönetim stratejilerinde ne denli etkin bir dönüşüm gerçekleştirdiğini göstermektedir.

Yenilenebilir Enerji Kullanımı Oranı:
Apple, 2011 yılında %0 oranında yenilenebilir enerji kullanırken, bu oran 2016 yılında %98’e, 2018 yılında ise %100’e ulaşmıştır. 2018’den itibaren tüm operasyonlarında %100 yenilenebilir enerji kullanan Apple, 2023 yılı itibarıyla bu başarıyı sürdürmüştür. Bu veriler, şirketin yalnızca kısa vadeli değil, uzun vadeli ve sürdürülebilir enerji yatırımları gerçekleştirdiğini doğrulamaktadır.

Emisyonlar:
Apple’ın operasyonel emisyonları (Scope 1 ve 2) 2011 yılında 10 ton CO₂e gibi çok düşük bir düzeyde başlamış; 2012–2015 yılları arasında enerji tüketiminin artmasına paralel olarak yaklaşık 10.000–16.000 ton CO₂e aralığında seyretmiştir. Ancak 2016 yılında başlatılan yenilenebilir enerji geçişi ile birlikte emisyonlar belirgin biçimde düşmüş, 2018 yılında yalnızca 400 ton CO₂e'ye gerilemiş ve 2019’dan itibaren sıfıra inmiştir. Bu sonuç, Apple’ın %100 yenilenebilir enerji hedefini yalnızca açıklamakla kalmayıp, veri ile desteklenebilir biçimde gerçekleştirdiğini ortaya koymaktadır.

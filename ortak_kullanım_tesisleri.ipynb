```python
import pandas as pd
import matplotlib.pyplot as plt

# Enerji ve emisyon verileri
data = {
    "Mali Yl": [
        2011, 2012, 2013, 2014, 2015, 2016, 2017,
        2018, 2019, 2020, 2021, 2022, 2023
    ],
    "Toplam Enerji Kullanm (kWh)": [
        42500, 38552300, 79462900, 108659700, 142615000, 145520900, 289195800,
        327663800, 339047649, 372901398, 384727076, 487921930, 483299062
    ],
    "Yenilenebilir Enerji Kullanm (%)": [
        0, 4, 59, 81, 85, 98, 99, 100, 100, 100, 100, 100, 100
    ],
    "Apple'n Emisyonlar (ton CO2e)": [
        10, 16500, 14500, 11000, 12700, 1600, 1500, 400, 0, 0, 0, 0, 0
    ]
}

df = pd.DataFrame(data)
df.set_index("Mali Yl", inplace=True)

# Grafik olutur
fig, ax1 = plt.subplots(figsize=(12, 6))

# Enerji kullanmn ubuk grafikle gster
ax1.bar(df.index, df["Toplam Enerji Kullanm (kWh)"], color='#A3BE8C', label='Toplam Enerji Kullanm (kWh)')
ax1.set_ylabel("Toplam Enerji (kWh)", color='#A3BE8C')
ax1.tick_params(axis='y', labelcolor='#A3BE8C')

# kincil y eksenine emisyonlar ekle
ax2 = ax1.twinx()
ax2.plot(df.index, df["Apple'n Emisyonlar (ton CO2e)"], color='#BF616A', marker='o', linewidth=2, label="Apple'n Emisyonlar (ton CO2e)")
ax2.set_ylabel("Apple'n Emisyonlar (ton CO2e)", color='#BF616A')
ax2.tick_params(axis='y', labelcolor='#BF616A')

# Balk ve gsterge
plt.title("Apple'n Enerji Kullanm ve Emisyon Eilimi (20112023)", fontsize=14)
fig.legend(loc='upper left', bbox_to_anchor=(0.15, 0.85))
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

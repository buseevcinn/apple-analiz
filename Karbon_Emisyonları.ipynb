1. Kapsamlara Göre Yıllık Toplam Emisyonların Yığılmış Çubuk Grafiği
   
Bu grafik, her yıl için toplam emisyonun Kapsam 1, Kapsam 2 ve Kapsam 3 arasında nasıl dağıldığını gösterir. Şirketin hangi kapsamda en fazla emisyon ürettiğini ve zaman içinde bu dağılımın nasıl değiştiğini anlamak için çok faydalıdır.
Toplam Emisyon Yapısı: Her yıl toplam emisyonun Kapsam 1, 2 ve 3 arasındaki oranını net bir şekilde gösterir.
Değişim Takibi: Yıllar geçtikçe farklı kapsamların toplam emisyondaki paylarının nasıl değiştiğini izlemeye olanak tanır.
Odak Alanları: Şirketin karbon ayak izini azaltma çabaları için hangi kapsamların öncelikli olması gerektiğini görsel olarak belirlemeye yardımcı olur.

```python
df_scope_yearly = df.groupby(['Yıl', 'Kapsam'])['Emisyon (kg CO2e)'].sum().unstack()

plt.figure(figsize=(12, 7))
df_scope_yearly.plot(kind='bar', stacked=True, figsize=(12, 7), colormap='viridis')
plt.title('Yıllara Göre Kapsam Bazında Toplam Emisyon Dağılımı', fontsize=16)
plt.xlabel('Yıl', fontsize=12)
plt.ylabel('Toplam Emisyon (kg CO2e)', fontsize=12)
plt.xticks(rotation=45, ha='right')

ax = plt.gca()
def custom_formatter_kgco2e(x, pos):
    if x >= 1e9:
        return f'{x*1e-9:.1f}B' # Milyarlar
    elif x >= 1e6:
        return f'{x*1e-6:.1f}M' # Milyonlar
    elif x >= 1e3:
        return f'{x*1e-3:.0f}K' # Binler
    else:
        return f'{x:.0f}'
ax.yaxis.set_major_formatter(mticker.FuncFormatter(custom_formatter_kgco2e))

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Kapsam', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

2. Belirli Emisyon Türlerinin Yıllara Göre Emisyon Değişimi (FacetGrid ile Çizgi Grafikleri)
   
Bu grafik, her bir Emisyon Türü için ayrı ayrı çizgi grafikleri oluşturarak, her bir türün zaman içindeki eğilimini daha detaylı incelemeyi sağlar. seaborn.FacetGrid kullanarak alt grafiklerin düzenini kolayca yönetebiliriz. "Üretim" ve "Ürün Kullanımı" gibi yüksek emisyonlu türlerin kendi logaritmik eksenlerine sahip olması için bu tür ayrı grafikler çok faydalıdır.
Detaylı Eğilim Analizi: Her bir emisyon türünün kendi özel eğilimini görmeyi sağlar. Örneğin, "Üretim" emisyonlarının sürekli arttığını veya "İş Seyahati" emisyonlarının pandeminin etkileriyle nasıl değiştiğini daha net görebilirsiniz.
Karşılaştırmalı Analiz: Aynı kapsam içinde veya farklı kapsamlar arasında belirli emisyon türlerinin performansını karşılaştırmak için kullanılabilir.
Ölçeklendirme Bağımsızlığı: sharey=False parametresi sayesinde her alt grafik kendi Y ekseni ölçeğini ayarlayabilir, bu da farklı büyüklükteki emisyon türlerini aynı anda anlamlı bir şekilde görselleştirmeyi sağlar.

```python
top_emission_types = df.groupby('Emisyon Türü')['Emisyon (kg CO2e)'].mean().nlargest(10).index

g = sns.FacetGrid(df[df['Emisyon Türü'].isin(top_emission_types)], col='Emisyon Türü', col_wrap=3, height=4, sharey=False)
g.map_dataframe(sns.lineplot, x='Yıl', y='Emisyon (kg CO2e)', marker='o')
g.set_axis_labels("Yıl", "Emisyon (kg CO2e)")
g.set_titles(col_template="{col_name}")

for ax in g.axes.flat:
    if df[(df['Emisyon Türü'] == ax.get_title()) & (df['Emisyon (kg CO2e)'] > 0)]['Emisyon (kg CO2e)'].min() > 0:
         ax.set_yscale('log')

    ax.yaxis.set_major_formatter(mticker.FuncFormatter(custom_formatter_kgco2e))
    ax.set_xticks(df['Yıl'].unique().astype(int))
    ax.grid(True, linestyle='--', alpha=0.7)

plt.suptitle('Seçili Emisyon Türlerinin Yıllara Göre Emisyon Değişimi', y=1.02, fontsize=18)
plt.tight_layout(rect=[0, 0, 1, 0.98])
plt.show()
```

3. Emisyon Türlerinin Yıllık Ortalamalarına Göre Sıralanmış Çubuk Grafiği
Bu grafik, tüm yıllar boyunca her bir Emisyon Türü'nün ortalama emisyonunu gösterir ve en yüksek ortalama emisyon türlerini vurgulayarak şirketin en büyük karbon kaynaklarını belirlemeye yardımcı olur.
Önceliklendirme: Şirketin en çok emisyon üreten alanlarını hızlıca belirlemeye yardımcı olur. Bu, emisyon azaltma stratejileri için öncelik alanlarını saptamak açısından kritik öneme sahiptir.
Genel Bakış: Belirli bir yılın dalgalanmalarından bağımsız olarak, her bir emisyon türünün genel etkisini gösterir.
Sıralama: Emisyon türlerini ortalama değerlerine göre sıralayarak, en kritik alanları görsel olarak vurgular.

```python
# 'Yıl' sütununu sıralama
df = df.sort_values(by=['Kapsam', 'Emisyon Türü', 'Yıl'])

# Her Emisyon Türü için ortalama emisyonu hesapla
df_avg_emission = df.groupby('Emisyon Türü')['Emisyon (kg CO2e)'].mean().sort_values(ascending=False).reset_index()

plt.figure(figsize=(14, 8))
# Düzeltme: `hue` parametresini ekleyerek ve `legend=False` yaparak uyarıyı gideriyoruz
sns.barplot(data=df_avg_emission, x='Emisyon (kg CO2e)', y='Emisyon Türü', hue='Emisyon Türü', palette='rocket', legend=False)
plt.title('Emisyon Türlerinin Yıllık Ortalama Emisyonlarına Göre Sıralaması (2019-2023)', fontsize=16)
plt.xlabel('Ortalama Emisyon (kg CO2e)', fontsize=12)
plt.ylabel('Emisyon Türü', fontsize=12)

# X ekseninde bilimsel notasyonu daha okunur hale getirme
ax = plt.gca()
def custom_formatter_kgco2e(x, pos):
    if x >= 1e9:
        return f'{x*1e-9:.1f}B' # Milyarlar
    elif x >= 1e6:
        return f'{x*1e-6:.1f}M' # Milyonlar
    elif x >= 1e3:
        return f'{x*1e-3:.0f}K' # Binler
    else:
        return f'{x:.0f}'
ax.xaxis.set_major_formatter(mticker.FuncFormatter(custom_formatter_kgco2e))

plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mticker # Y ekseni formatlaması için
data = {
    'Su Kaynağı': ['Tatlı su', 'Tatlı su', 'Tatlı su', 'Tatlı su', 'Tatlı su',
                   'Geri Dönüştürülmüş Su', 'Geri Dönüştürülmüş Su', 'Geri Dönüştürülmüş Su', 'Geri Dönüştürülmüş Su', 'Geri Dönüştürülmüş Su',
                   'Alternatif Kaynaklar', 'Alternatif Kaynaklar', 'Alternatif Kaynaklar', 'Alternatif Kaynaklar', 'Alternatif Kaynaklar'],
    'Yıl': [2023, 2022, 2021, 2020, 2019,
            2023, 2022, 2021, 2020, 2019,
            2023, 2022, 2021, 2020, 2019],
    'Tüketim (milyon galon)': [1411, 1380, 1259, 1168, 1178,
                               151, 142, 141, 113, 106,
                               48, 5, 7, 5, 7]
}
df_water = pd.DataFrame(data)
df_water = df_water.sort_values(by=['Yıl', 'Su Kaynağı'])
def custom_formatter_million_gallon(x, pos):
    if x >= 1000: # 1000 milyon galon = 1 milyar galon
        return f'{x*1e-3:.1f}B' # Bilyonlar
    elif x >= 1:
        return f'{x:.0f}'
    return f'{x:.0f}' # Küçük değerler için
plt.figure(figsize=(12, 7))
sns.lineplot(data=df_water, x='Yıl', y='Tüketim (milyon galon)', hue='Su Kaynağı',
             marker='o', markersize=8, linewidth=2.5, palette='tab10') # Farklı renkler ve belirgin markerlar
plt.title('Yıllara Göre Su Kaynakları Tüketim Değişimi', fontsize=16, fontweight='bold')
plt.xlabel('Yıl', fontsize=12)
plt.ylabel('Tüketim (Milyon Galon)', fontsize=12)
plt.xticks(df_water['Yıl'].unique().astype(int))
ax = plt.gca() 
ax.yaxis.set_major_formatter(mticker.FuncFormatter(custom_formatter_million_gallon))
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Su Kaynağı', bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mticker # Y ekseni formatlamas iin
data = {
    'Su Kayna': ['Tatl su', 'Tatl su', 'Tatl su', 'Tatl su', 'Tatl su',
                   'Geri Dntrlm Su', 'Geri Dntrlm Su', 'Geri Dntrlm Su', 'Geri Dntrlm Su', 'Geri Dntrlm Su',
                   'Alternatif Kaynaklar', 'Alternatif Kaynaklar', 'Alternatif Kaynaklar', 'Alternatif Kaynaklar', 'Alternatif Kaynaklar'],
    'Yl': [2023, 2022, 2021, 2020, 2019,
            2023, 2022, 2021, 2020, 2019,
            2023, 2022, 2021, 2020, 2019],
    'Tketim (milyon galon)': [1411, 1380, 1259, 1168, 1178,
                               151, 142, 141, 113, 106,
                               48, 5, 7, 5, 7]
}
df_water = pd.DataFrame(data)
df_water = df_water.sort_values(by=['Yl', 'Su Kayna'])
def custom_formatter_million_gallon(x, pos):
    if x >= 1000: # 1000 milyon galon = 1 milyar galon
        return f'{x*1e-3:.1f}B' # Bilyonlar
    elif x >= 1:
        return f'{x:.0f}'
    return f'{x:.0f}' # Kk deerler iin
plt.figure(figsize=(12, 7))
sns.lineplot(data=df_water, x='Yl', y='Tketim (milyon galon)', hue='Su Kayna',
             marker='o', markersize=8, linewidth=2.5, palette='tab10') # Farkl renkler ve belirgin markerlar
plt.title('Yllara Gre Su Kaynaklar Tketim Deiimi', fontsize=16, fontweight='bold')
plt.xlabel('Yl', fontsize=12)
plt.ylabel('Tketim (Milyon Galon)', fontsize=12)
plt.xticks(df_water['Yl'].unique().astype(int))
ax = plt.gca() 
ax.yaxis.set_major_formatter(mticker.FuncFormatter(custom_formatter_million_gallon))
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Su Kayna', bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.tight_layout()
plt.show()
```

Çizgi Grafiği - Zaman Serisi Karşılaştırması
Hem tesis bazında, hem tasarruf tipi bazında yıllara göre değişimi izlemek için

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Örnek veri yapısı
df = pd.DataFrame({
    "Yıl": [2019,2020,2021,2022,2023]*4,
    "Tesis": ["Kurumsal Tesis"]*5 + ["Kurumsal Tesis"]*5 + ["Tedarikçi Tesisi"]*5 + ["Tedarikçi Tesisi"]*5,
    "Tasarruf Tipi": ["Elektrik Tasarrufu"]*5 + ["Yakıt Tasarrufu"]*5 + ["Elektrik Tasarrufu"]*5 + ["Yakıt Tasarrufu"]*5,
    "Birim": ["MWh/year"]*10 + ["MWh/year"]*5 + ["MMBtu/year"]*5,
    "Tasarruf": [
        118830, 215260, 223940, 290220, 298500,    # Kurumsal Elektrik
        162280, 136820, 110330, 116190, 126540,    # Kurumsal Yakıt
        943890, 1101440, 1418825350, 1620425230, 2040000000, # Tedarikçi Elektrik
        25120, 752690, 1047440, 2038930, 2261060   # Tedarikçi Yakıt
    ]
})

# Grafik
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Yıl", y="Tasarruf", hue="Tesis", style="Tasarruf Tipi", markers=True, dashes=False)
plt.yscale("log")  # Çok yüksek rakam farkı olduğu için log kullanmak uygun olur.
plt.title("Yıllara Göre Enerji Tasarrufu (Kurumsal vs Tedarikçi)")
plt.ylabel("Tasarruf (log ölçek)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
